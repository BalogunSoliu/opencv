```python
import cv2 as cv
import numpy as np
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
mainfolder='pictures'
folders=os.listdir(mainfolder)
allfiles=[]
fid=[]
for folder in folders:
    subf=os.path.join(mainfolder,folder)
    files=os.listdir(subf)
    for file in files:
        filename=os.path.join(subf,file)
        allfiles.append(filename)
        fid.append(folder)
```

```python
for picture in index:
    findex = np.random.choice(index)
    img = cv.imread(allfiles[findex])
    img=cv.resize(img, (249,350))
    sindex = np.random.choice(index)
    template = cv.imread(allfiles[np.random.choice(index)])
    template=cv.resize(template, (249,350))
    for method in methods:
        res = cv.matchTemplate(img,template,method)
    
        if res>.5 or res == 0.0:
             d=f'similar:  {res}'
             pair.append([findex,sindex])
        else:
            d=f'not similar:  {res}'
    plt.subplot(122),plt.imshow(img),plt.title(d)
    plt.subplot(121),plt.imshow(template),plt.title(d)
    plt.show()
    
```
